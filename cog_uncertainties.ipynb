### notebook to calculate uncertainties on curve of growth fits

```python
from astropy.table import Table, vstack
import numpy as np
import tqdm
import matplotlib.pyplot as plt
import matplotlib as mpl

from IPython.display import Image
import os
```

```python
DATA_FOLDER = '/pscratch/sd/n/nravi/GV_classification/'
COV_FOLDER = DATA_FOLDER + 'cov_matrices/'
master_table_fn = DATA_FOLDER + 'nsa_v1_0_1_cd_v2.fits'
```

```python
master_table = Table.read(master_table_fn)
master_table[:5]
```

```python
master_table_dict = {}
for i in tqdm.tqdm(range(len(master_table))):
    master_table_dict[master_table['IAUNAME'][i]] = i
```

```python
# make list of files in cov_matrices

cov_files = os.listdir('/pscratch/sd/n/nravi/GV_classification/cov_matrices')
```

```python
# create cols in nsa
master_table['g_mtot_err'] = np.ones(len(master_table)) * np.nan
master_table['g_m0_err'] = np.ones(len(master_table)) * np.nan
master_table['g_a1_err'] = np.ones(len(master_table)) * np.nan
master_table['g_a2_err'] = np.ones(len(master_table)) * np.nan
master_table['i_mtot_err'] = np.ones(len(master_table)) * np.nan
master_table['i_m0_err'] = np.ones(len(master_table)) * np.nan
master_table['i_a1_err'] = np.ones(len(master_table)) * np.nan
master_table['i_a2_err'] = np.ones(len(master_table)) * np.nan
master_table['NSA_cd_err'] = np.ones(len(master_table)) * np.nan
```

```python
for i in tqdm.tqdm(range(len(cov_files))):

    fn = cov_files[i]
    iauname, _ = fn.split('_', 1)
    idx = master_table_dict[iauname]

    #open cov matrix

    cov = np.load('/pscratch/sd/n/nravi/GV_classification/cov_matrices/' + fn)
    err = np.sqrt(np.diag(cov))

    if '_g_' in fn:
    
        master_table['g_mtot_err'][idx] = err[0]
        master_table['g_m0_err'][idx] = err[1]
        master_table['g_a1_err'][idx] = err[2]
        master_table['g_a2_err'][idx] = err[3]

    elif '_i_' in fn:
        master_table['i_mtot_err'][idx] = err[0]
        master_table['i_m0_err'][idx] = err[1]
        master_table['i_a1_err'][idx] = err[2]
        master_table['i_a2_err'][idx] = err[3]
    
```

```python
master_table.remove_column('NSA_cd_err')
```

```python
master_table.write(master_table_fn, overwrite=True)
```
