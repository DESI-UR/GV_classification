Compare the photometry in the KIAS-VAGC and in the NSA (both versions 0.1.2 and 1.0.1).

```python
from astropy.table import Table

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Catalogs

```python
data_directory = '/Users/kellydouglass/Documents/Research/data/'

NSA_v012_filename = 'nsa_v0_1_2_VAGC.fits'
NSA_v101_filename = 'nsa_v1_0_1_VAGC.fits'
```

```python
NSA_v012 = Table.read(data_directory + NSA_v012_filename, format='fits')
NSA_v101 = Table.read(data_directory + NSA_v101_filename, format='fits')
```

```python
NSA_v012['u_r'] = NSA_v012['ABSMAG'][:,2] - NSA_v012['ABSMAG'][:,4]
NSA_v101['u_r'] = NSA_v101['ELPETRO_ABSMAG'][:,2] - NSA_v101['ELPETRO_ABSMAG'][:,4]
```

# Color

```python
plt.figure(figsize=(12,5))

plt.subplot(121)
plt.hist2d(NSA_v012['u_r'], NSA_v012['u_r_KIAS'], bins=100, range=[[-1,5], [-1,5]], norm=mpl.colors.LogNorm())
plt.plot([-1,5], [-1,5], ':')
plt.title('NSA v0.1.2')
plt.xlabel('u-r (NSA)')
plt.ylabel('u-r (KIAS)')
plt.xlim([-1,5])
plt.ylim([-1,5])

plt.subplot(122)
plt.hist2d(NSA_v101['u_r'], NSA_v101['u_r_KIAS'], bins=100, range=[[-1,5], [-1,5]], norm=mpl.colors.LogNorm())
plt.plot([-1,5], [-1,5], ':')
plt.title('NSA v1.0.1')
plt.xlabel('u-r (NSA)')
#plt.ylabel('u-r (KIAS)')
plt.xlim([-1,5])
plt.ylim([-1,5])

plt.tight_layout();
```

# Magnitude

```python
plt.figure(figsize=(12,5))

plt.subplot(121)
plt.hist2d(np.log10(NSA_v012['RNMGY'][:,4]), 
           NSA_v012['prmag'], 
           bins=100, 
           range=[[0,11], [10,18]], 
           norm=mpl.colors.LogNorm())
#plt.plot([0,11], [10,18], ':')
plt.title('NSA v0.1.2')
plt.xlabel('log(reconstructed AB nanomaggies) (NSA)')
plt.ylabel('prmag (KIAS)')
plt.xlim([0,11])
plt.ylim([10,18])

plt.subplot(122)
plt.hist2d(np.log10(NSA_v101['ELPETRO_RNMGY'][:,4])[np.isfinite(np.log10(NSA_v101['ELPETRO_RNMGY'][:,4]))], 
           NSA_v101['u_r_KIAS'][np.isfinite(np.log10(NSA_v101['ELPETRO_RNMGY'][:,4]))], 
           bins=[100,100], 
           range=[[0,4], [10,18]], 
           norm=mpl.colors.LogNorm())
#plt.plot([0,11], [10,18], ':')
plt.title('NSA v1.0.1')
plt.xlabel('log(elliptical Petrosian reconstructed AB nanomaggies) (NSA)')
plt.ylabel('prmag (KIAS)')
plt.xlim([0,4])
plt.ylim([10,18])

plt.tight_layout();
```
