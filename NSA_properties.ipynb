```python
from astropy.table import Table
from astropy.io import fits

import numpy as np

import matplotlib.pyplot as plt
```

# Data

```python
NSAcolumns_needed = ['RA', 'DEC', 'Z', 'u_r', 'u_r_KIAS', 'NUV_r', 'CMD_class', 'ELPETRO_RNMGY', 'ELPETRO_ABSMAG']
KIAScolumns_needed = ['ra', 'dec', 'redshift', 'u_r', 'NUV_r', 'CMD_class']


#v012_filename = '/Users/kellydouglass/Documents/Research/data/kias1033_5_P-MJD-F_MPAJHU_ZdustOS_NSAv012_CMDJan2020.txt'
v012_filename = '/Users/kellydouglass/Documents/Research/data/nsa_v0_1_2_VAGC.fits'

# NSAv012 = Table.read(v012_filename, 
#                      format='ascii.commented_header', 
#                      include_names=NSAcolumns_needed)
hdul = fits.open(v012_filename)
NSAv012 = Table([hdul[1].data.RA, 
                 hdul[1].data.DEC, 
                 hdul[1].data.Z, 
                 hdul[1].data.u_r, 
                 hdul[1].data.NUV_r, 
                 hdul[1].data.RNMGY, 
                 hdul[1].data.ABSMAG],
                names=['RA', 'DEC', 'Z', 'u_r', 'NUV_r', 'RNMGY', 'ABSMAG'])
hdul.close()


#v101_filename = '/Users/kellydouglass/Documents/Research/data/kias1033_5_P-MJD-F_MPAJHU_ZdustOS_NSAv101_CMDJan2020.txt'
v101_filename = '/Users/kellydouglass/Documents/Research/data/NSA_v1_0_1_VAGC_CMDJan2020_vflag-V2-VF.fits'

# NSAv101 = Table.read(v101_filename, 
#                      format='fits',
#                      #format='ascii.commented_header', 
#                      include_names=NSAcolumns_needed)
hdul = fits.open(v101_filename)
NSAv101 = Table([hdul[1].data.RA, 
                 hdul[1].data.DEC, 
                 hdul[1].data.Z, 
                 hdul[1].data.u_r, 
                 hdul[1].data.u_r_KIAS,
                 hdul[1].data.NUV_r, 
                 hdul[1].data.CMD_class, 
                 hdul[1].data.ELPETRO_RNMGY, 
                 hdul[1].data.ELPETRO_ABSMAG], 
                names=NSAcolumns_needed)
hdul.close()


kias_filename = '/Users/kellydouglass/Documents/Research/data/kias1033_5_MPAJHU_ZdustOS.txt'

kias = Table.read(kias_filename, 
                  format='ascii.commented_header', 
                  include_names=KIAScolumns_needed)
```

## Calculate alternate colors

```python
NSAv012['NUV_r_kcorr'] = -2.5*np.log10(NSAv012['RNMGY'][:,1]/NSAv012['RNMGY'][:,4])
NSAv101['NUV_r_kcorr'] = -2.5*np.log10(NSAv101['ELPETRO_RNMGY'][:,1]/NSAv101['ELPETRO_RNMGY'][:,4])

NSAv012['NUV_r_absmag'] = NSAv012['ABSMAG'][:,1] - NSAv012['ABSMAG'][:,4]
NSAv101['NUV_r_absmag'] = NSAv101['ELPETRO_ABSMAG'][:,1] - NSAv101['ELPETRO_ABSMAG'][:,4]
```

```python
NSAv012['u_r_kcorr'] = -2.5*np.log10(NSAv012['RNMGY'][:,2]/NSAv012['RNMGY'][:,4])
NSAv101['u_r_kcorr'] = -2.5*np.log10(NSAv101['ELPETRO_RNMGY'][:,2]/NSAv101['ELPETRO_RNMGY'][:,4])

NSAv012['u_r_absmag'] = NSAv012['ABSMAG'][:,2] - NSAv012['ABSMAG'][:,4]
NSAv101['u_r_absmag'] = NSAv101['ELPETRO_ABSMAG'][:,2] - NSAv101['ELPETRO_ABSMAG'][:,4]
```

```python
plt.figure(figsize=(12,5))


plt.subplot(121)

plt.plot(NSAv012['NUV_r'], NSAv012['NUV_r_absmag'], '.', alpha=0.1, label='v012')
plt.plot(NSAv101['NUV_r'], NSAv101['NUV_r_absmag'], '.', alpha=0.1, label='v101')

plt.plot([-10, 40], [-10, 40], 'r')

plt.xlabel('NUV - r from (ELPETRO_)FLUX')
plt.ylabel('NUV - r from (ELPETRO_)ABSMAG')

plt.xlim([-5,30])

plt.legend()


plt.subplot(122)

plt.plot(NSAv012['NUV_r_kcorr'], NSAv012['NUV_r_absmag'], '.', alpha=0.1, label='v012')
plt.plot(NSAv101['NUV_r_kcorr'], NSAv101['NUV_r_absmag'], '.', alpha=0.1, label='v101')

plt.plot([-10, 40], [-10, 40], 'r')

plt.xlabel('NUV - r from (ELPETRO_)RNMGY')
plt.ylabel('NUV - r from (ELPETRO_)ABSMAG')

plt.xlim([-1,9])

plt.legend();
```

```python
plt.figure(figsize=(12,5))


plt.subplot(121)

plt.plot(NSAv012['u_r'], NSAv012['u_r_absmag'], '.', alpha=0.1, label='v012')
plt.plot(NSAv101['u_r'], NSAv101['u_r_absmag'], '.', alpha=0.1, label='v101')

plt.plot([-10, 40], [-10, 40], 'r')

plt.xlabel('u - r from (ELPETRO_)FLUX')
plt.ylabel('u - r from (ELPETRO_)ABSMAG')

plt.xlim([-5,10])
plt.ylim([-5,10])

plt.legend()


plt.subplot(122)

plt.plot(NSAv012['u_r_kcorr'], NSAv012['u_r_absmag'], '.', alpha=0.1, label='v012')
plt.plot(NSAv101['u_r_kcorr'], NSAv101['u_r_absmag'], '.', alpha=0.1, label='v101')

plt.plot([-10, 40], [-10, 40], 'r')

plt.xlabel('u - r from (ELPETRO_)RNMGY')
plt.ylabel('u - r from (ELPETRO_)ABSMAG')

plt.xlim([-1,5])
plt.ylim([-1,5])

plt.legend();
```
